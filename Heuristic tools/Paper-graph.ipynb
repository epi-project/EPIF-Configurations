```python
import pandas as pd
import matplotlib.pyplot as plt

# creating dataframe
dataFrame = pd.DataFrame({
   "f": ['Encrypt', 'Firewall', 'DPI', 'Available'],"per": [10, 20, 20, 50]
})
colors = ['pink', 'silver', 'steelblue', 'limegreen']
explode = (0.05, 0.05, 0.05, 0.05)
# plot a Pie Chart for Registration Price column with label Car column
plt.pie(dataFrame["per"], labels = dataFrame["f"], autopct='%1.0f%%',
  colors=colors, explode=explode)
plt.show()
```
